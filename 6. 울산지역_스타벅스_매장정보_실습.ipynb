## 시작 전 사전지식
### 문자열에서 일부의 문자열을 추출하는 슬라이싱
* 문자열에서 일부의 문자열을 추출하는 것을 슬라이싱이라 한다.
* 문법 : 문자열[시작위치:마지막 위치]
* 문자열의 시작위치에서 마지막 위치 전까지 문자열 잘라낸다.
* 시작위치는 항상 0부터 시작한다.  
  

> name = 'hello python'    
> name[0:4] -> 'hell'  
> name[2:6] -> 'llo '  
> name[0:-1] -> 'hello pytho'  
> name[0:-2] -> 'hello pyth'  
> name[:3] -> 'hel' : 시작 위치를 생략하면 처음부터 시작    
> name[6:] -> 'python' : 마지막 위치를 생략하면 문자열 끝까지!

### 4단계 : 본격적인 스크래핑 시작!

```python
# 지역 검색 버튼을 찾는 코드를 작성
loca = driver.find_element(By.CLASS_NAME, 'loca_search')
                     
# 찾은 버튼을 클릭하는 코드를 작성
loca.click()

#3초간 대기
time.sleep(3)
```

```python
# 구/군 정보가 위치한 태그를 찾는 코드 작성
gugun = driver.find_element(By.CLASS_NAME, 'gugun_arae_box')

# 구/군 정보가 위치한 태그 안에서 전체 구군에 대한 li 태그를 찾는 코드 작성
total_gugun = gugun.find_elements(By.TAG_NAME, 'li')[0]

# 찾은 태그 클릭
total_gugun.click()

# 3초간 대기
time.sleep(3)
```
